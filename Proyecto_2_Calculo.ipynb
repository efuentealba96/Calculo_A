**Modelo 2  que utilizaremos sera el de Kosmeyer Peppas:**

**Funcion con una variable**

Para dar un inicio a la optimización, primero veeremos una funcion univariable que cumple con que su derivada es igual a 0 y su segunda derivada sea distinta de cero, siendo un positivo en caso de un mínimo.    Estas condiciones se les reconoce como **condición necesaria** (primera derivada igual a cero), y **condición suficiente** (segunda derivada distinta de cero)

La funcion en cuestión es la siguiente:

$F(x)=x^2+3x+5$

```python
def Fx(x):
  return x**2 + 3*x + 5

def PrimeraDerivada(x):
   return 2*x + 3

def SegundaDerivada(x):
  return 2

thisX = range(-10, 11)

print("\nFuncion Principal")
plt.plot(thisX, [Fx(i) for i in thisX])
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)
plt.show()
print("\nPrimera Derivada")
plt.plot(thisX, [PrimeraDerivada(i) for i in thisX])
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)
plt.show()
print("\nSegunda Derivada")
plt.plot(thisX, [SegundaDerivada(i) for i in thisX])
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)
plt.show()
```

**Funcion con dos variables**

Para el caso de una función de dos variables, su **condición necesaria** para obtener un óptimo es que su **grandiente** debe estar igualada a cero para cada uno de sus componentes, y para su **condición suficiente**, es necesario que la **Matriz Hessiana** es definida, que para el caso de un mínimo, ésta debe ser positiva.

A continuación, se realizarán los pasos correspondientes para verificar lo anterior, para la cual haremos uso de la libreria sympy.

```python
from __future__ import division
import sympy as sym
from sympy.matrices import hessian, zeros
sym.init_printing(use_latex="mathjax")

def gradient(f, varls):
    # Calcula el gradiente de una función f(x1,x2,...)
    n = len(varls)
    G = zeros(n,1)
    for i in range(n):
        G[i] = f.diff(varls[i])
    return G
```

Definimos las variables simbólicas a utilizar y la función $A(x1,x2)$

```python
x1,x2,V = sym.symbols("x_1,x_2,V")
A = 2*(V/x2) + 2*(V/x1) + x1*x2
```

Calculamos el gradiente de la función y establecemos el sistema de ecuaciones:

```python
G = gradient(A.subs(V,0.08), (x1,x2))
sym.Eq(G,zeros(2,1))
```

Resolviendo el sistema de ecuaciones, se tiene:

```python
sol = sym.solve(sym.Eq(G,zeros(2,1)), (x1,x2))
sol
```

Calculando la matriz Hessiana:

```python
H = hessian(A, (x1,x2))
# evaluando en el punto calculado
_x1, _x2 = sol[0][0], sol[0][1]
H = H.subs({x1:_x1, x2:_x2, V:0.08})
H
```

Calculando los eigenvalores:

```python
egvals = H.eigenvals()
egvals
```

Dado que tanto $λ1=1$ como $λ2=3$ son positivos, entonces el punto calculado es un mínimo global no restringido.

Ahora, si calculamos el valor de la altura $x3$:

```python
_x3 = 0.08/(_x1*_x2)
_x3
```

**Utilizando SciPy**

```python
from scipy.optimize import minimize

def A(x):
    v = 0.08
    return 2*(v/x[1]) + 2*(v/x[0]) + x[0]*x[1]
    
def dxA(x):
    v = 0.08
    dfdx0 = -2*(v/x[0]**2) + x[1]
    dfdx1 = -2*(v/x[1]**2) + x[0]
    return np.array([dfdx0, dfdx1])

cons = ({'type': 'ineq',
        'fun' : lambda x: np.array([x[0] - 0.1]),
        'jac' : lambda x: np.array([1,0])},
        {'type': 'ineq',
        'fun' : lambda x: np.array([x[1] - 0.1]),
        'jac' : lambda x: np.array([0,1])},
        {'type': 'ineq',
        'fun' : lambda x: np.array([0.08/(x[0]*x[1]) - 0.6]),
        'jac' : lambda x: np.array([-0.08/(x[0]**2*x[1]), -0.08/(x[0]*x[1]**2)])},
        {'type': 'ineq',
        'fun' : lambda x: -np.array([0.08/(x[0]*x[1]) - 0.8]),
        'jac' : lambda x: -np.array([-0.08/(x[0]**2*x[1]), -0.08/(x[0]*x[1]**2)])},
       )
        
#~ cons = ({'type': 'ineq',
        #~ 'fun' : lambda x: np.array([x[0] - 0.1])},
        #~ {'type': 'ineq',
        #~ 'fun' : lambda x: np.array([x[1] - 0.1])},
        #~ {'type': 'ineq',
        #~ 'fun' : lambda x: np.array([0.08/(x[0]*x[1]) - 0.6])}
        #~ )
    
res = minimize(A, [0.3,0.3], method='SLSQP', jac= dxA, 
               constraints=cons)
res
```
